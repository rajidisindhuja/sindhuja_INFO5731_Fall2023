Question 2 (10 points): Write python code to collect 1000 data samples you discussed above.

```python
#import necessary packages
import random
import csv

# Generate synthetic data for robot deliveries
def generate_robot_delivery_data():
    data = []
    for _ in range(1000):
        delivery_date = f"{random.randint(2020, 2022)}-{random.randint(1, 12)}-{random.randint(1, 28)}"
        delivery_distance = random.uniform(0.1, 10.0)  # Kilometers
        delivery_time = random.uniform(5, 60)  # Minutes
        goods_type = random.choice(['Food', 'Packages', 'Medicine', 'Groceries'])
        data.append([delivery_date, delivery_distance, delivery_time, goods_type])
        # Print the generated data
        print(f"Delivery Date: {delivery_date}, Distance: {delivery_distance} km, Time: {delivery_time} min, Goods Type: {goods_type}")

    return data

# Save the generated data to a CSV file
def save_data_to_csv(data, filename):
    with open(filename, 'w', newline='') as file:
        writer = csv.writer(file)
        writer.writerow(['Delivery Date', 'Delivery Distance (km)', 'Delivery Time (min)', 'Goods Type'])
        writer.writerows(data)

if __name__ == '__main__':
    robot_delivery_data = generate_robot_delivery_data()
    save_data_to_csv(robot_delivery_data, 'robot_delivery_data.csv')



```

Question 3 (10 points): Write python code to collect 1000 articles from Google Scholar (https://scholar.google.com/), Microsoft Academic (https://academic.microsoft.com/home), or CiteSeerX (https://citeseerx.ist.psu.edu/index), or Semantic Scholar (https://www.semanticscholar.org/), or ACM Digital Libraries (https://dl.acm.org/) with the keyword "information retrieval". The articles should be published in the last 10 years (2013-2023).

The following information of the article needs to be collected:

(1) Title

(2) Venue/journal/conference being published

(3) Year

(4) Authors

(5) Abstract

```python

import requests
from bs4 import BeautifulSoup

def webscrape_schol_articles(url):
    response = requests.get(url)  # here we are sending a GET request to the URL
    soup = BeautifulSoup(response.text, 'html.parser')   # here we are parsing the HTML content of the page using beautiful soup
    articles = soup.find_all('div', class_='gs_ri')  # here we are locating and gathering data for each article on the page.

  # Loop through the articles and extract the desired information
    for i in articles:
        title = i.find('h3', class_='gs_rt').text
        venue = i.find('div', class_='gs_a').text.split(' - ')[-1]
        year = i.find('div', class_='gs_a').text.split(' - ')[0]
        authors = i.find('div', class_='gs_a').text.split(' - ')[1]
        abstract = i.find('div', class_='gs_rs').text

        print("Title:", title)
        print("Venue:", venue)
        print("Year:", year)
        print("Authors:", authors)
        print("Abstract:", abstract)
        print("\n---\n")

# Loop through plenty of pages of results to get 1,000 articles.
articles_per_page = 10  # A page of Google Scholar normally shows 10 items.
total_articles = 1000

for page in range(0, total_articles, articles_per_page):
  # URL of Google Scholar search results
    page_url = f"https://scholar.google.com/scholar?hl=en&as_sdt=0%2C5&q=information+retrieval&as_ylo=2013&as_yhi=2023&start={page}"
    webscrape_schol_articles(page_url)
```
